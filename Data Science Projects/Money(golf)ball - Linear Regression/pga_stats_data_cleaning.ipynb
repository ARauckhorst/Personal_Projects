```python
import pga_stats_scraper as pga
import pandas as pd
import numpy as np
```

```python
df = pd.concat([pga.df2015, pga.df2016, pga.df2017, pga.df2018, pga.df2019])
```

```python
df
```

```python
df.info()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.rename(columns = {'index':'player'}, inplace = True)
```

```python
df.head()
```

```python
df = df.astype({'scoring_avg': 'float',
           'driving_dist': 'float',
           'driving_acc': 'float',
           'gir_fairway': 'float',
           'gir_other': 'float',
           'sg_tee_to_green': 'float',
           'sg_total': 'float',
           'sg_putting': 'float',
           'scrambling': 'float',
           '3_putt_avoid': 'float',
           'putts_per_round': 'float',
           'bogey_avoidance': 'float'})
```

```python
df['earnings'] = df['earnings'].str.replace('$', '')
df['earnings'] = df['earnings'].str.replace(',', '')
df['earnings'] = df['earnings'].astype(int)
```

```python
df.head()
```

```python
def parse_ht(ht):
    # format: 7' 2"
    ht = ht.split("' ")
    ft = float(ht[0])
    inches = float(ht[1].replace("\"",""))
    return (12 * ft) + inches

df['prox_to_hole'] = df["prox_to_hole"].apply(lambda x: parse_ht(x))
```
